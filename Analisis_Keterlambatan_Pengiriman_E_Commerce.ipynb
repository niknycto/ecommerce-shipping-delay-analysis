````python
import numpy as np
import pandas as pd

# Menentukan volume data
num_rows = 10999
num_columns = 12
data = np.random.rand(num_rows, num_columns)
df = pd.DataFrame(data, columns=[f'Feature_{i}' for i in range(1, num_columns + 1)])

# Menentukan volume data
def calculate_data_volume(dataframe):
    # Menghitung jumlah baris dan kolom
    num_rows, num_columns = dataframe.shape

    # Menghitung ukuran satu baris data dalam bytes
    size_per_row = dataframe.memory_usage(deep=True).sum() / num_rows

    # Menghitung total volume data dalam megabytes (MB)
    total_volume_mb = (size_per_row * num_rows) / (1024 ** 2)

    return total_volume_mb

# Menampilkan volume data
data_volume = calculate_data_volume(df)
print(f'Volume data: {data_volume:.2f} MB')
````

````python
from scipy.stats import norm

# Menentukan sample
p_estimasi = 0.5 # Proporsi yang diestimasi dalam populasi
tingkat_kepercayaan = 0.95 # Tingkat kepercayaan 95%
margin_error = 0.03 # Margin of error 3%
sigma = np.sqrt((p_estimasi * (1 - p_estimasi)) / 10999) # Menghitung deviasi standar dari proporsi estimasi
z_score = norm.ppf((1 + tingkat_kepercayaan) / 2) # Menghitung skor z berdasarkan tingkat kepercayaan

# Menghitung ukuran sampel yang diperlukan
ukuran_sampel = int((z_score**2 * p_estimasi * (1 - p_estimasi)) / margin_error**2)

print(f"Ukuran sampel yang diperlukan: {ukuran_sampel}")
````

````python
# Melihat Outlier
# Menentukan ukuran layout subplot sesuai dengan jumlah kolom
num_cols = len(df1.columns)
num_rows = (num_cols // 4) + 1 if num_cols % 4 != 0 else num_cols // 4

# Membuat box plot
plt.figure(figsize=(10, 5 * num_rows))

# Loop melalui setiap kolom
for i, column in enumerate(df1.columns, 1):
    plt.subplot(num_rows, 4, i)
    plt.boxplot(df1[column].dropna())
    plt.title('Box Plot untuk {}'.format(column), fontsize=10)

# Menyusun tata letak
plt.tight_layout()

# Menampilkan semua box plot
plt.show()
````

**Keterangan :** Berdasarkan visualisasi boxplot, Outliers yang berjumlah 357 baris tersebar pada kolom Prior_purchases dan kolom Discount_offered. pada Prior_purchases terdapat outliers yang signifikan di atas distribusi data. Sedangkan pada Discount_offered menunjukan bahwa cukup banyak nilai tingkat ekstreem yang serupa.

````python
# Menangani Outlier
numeric_features = df1.select_dtypes(include=['float64', 'int64'])

# Fungsi untuk melakukan winsorizing pada kolom
def winsorize_column(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    column = np.where(column < lower_bound, lower_bound, column)
    column = np.where(column > upper_bound, upper_bound, column)
    return column

# Melakukan winsorizing untuk setiap kolom numerik
numeric_features = numeric_features.apply(winsorize_column)

# Memperbarui DataFrame asli dengan hasil winsorizing
df1.update(numeric_features)
````

**Keterangan :** Unutk menangani outliers, dapat menggunakan berbagai teknik seperti menghapus outlier, transformasi data, winsoriing, dan Interquartil Range. kali ini menggunakan teknik Winsorizing.

Winsorizing adalah suatu teknik dalam statistika yang digunakan untuk menangani outliers pada dataset. Teknik ini melibatkan penggantian nilai-nilai ekstrem (outliers) dengan nilai yang berada di batas atas atau batas bawah yang masih dianggap wajar. Dengan kata lain, nilai-nilai ekstrem tersebut "dihapus" dan digantikan oleh nilai terdekat yang tidak dianggap sebagai outliers.

**Keterangan :** Melakukan rekayasa fitur dengan mengelompokkan fitur independen sebagai X, dan dependen sebagai y.
````python
# Melakukan Over Sampling untuk memaksimalkan model
from imblearn.over_sampling import RandomOverSampler

rus = RandomOverSampler(sampling_strategy = 1)
X_res_over, y_res_over = rus.fit_resample(X, y)
````

**Keterangan :** Melakukan over-sampling sebelum pemodelan adalah untuk meningkatkan peforma model, menangani bias model, meningkatkan konsistensi dan reliabilitas, dan mengoptimalkan penggunaan data yang tersedia.

**Keterangan :** Fungsi melakukan transformasi data menggunakan `StandardScaler` adalah untuk menghasilkan data yang memiliki mean (rata-rata) 0 dan standar deviasi (deviasi standar) 1. Transformasi ini memastikan bahwa setiap fitur memiliki skala yang seragam, yang dapat membantu algoritma machine learning yang sensitif terhadap perbedaan skala antar fitur. Dengan demikian, fitur-fitur dalam data akan berada pada skala yang serupa, mengoptimalkan kinerja model dan memfasilitasi interpretasi hasil analisis.
